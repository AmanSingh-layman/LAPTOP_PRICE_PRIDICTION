```python
def fun(row):
    if row['Gen'] in ['13th Gen', '12th Gen', '11th Gen', '10th Gen', '9th Gen', '8th Gen', '7th Gen', '6th Gen', '5th Gen', '4th Gen', '3rd Gen']:
        return row['Gen']
    else:
        if row['Gen'] == 'intel Celeron' or row['Gen'] == 'Intel Celeron':
            return '6th Gen'
        elif row['Gen'] == 'Apple M2':
            return '13th Gen'
        elif row['Gen'] == 'Apple M1' or row['Gen'] == 'MediaTek MTK8788':
            return '11th Gen'
        elif row['Gen'] == 'Intel Core':
            return '12th Gen'
        elif row['Gen'] == 'Intel Pentium' or row['Gen'] == 'AMD Athlon':
            return '7th Gen'
        elif row['Gen'] == 'Intel Atom' or row['Gen'] == 'Intel':
            return '3rd Gen'
        else:
            return '10th Gen'

df1['Gen_pro'] = df1.apply(fun, axis=1)
df1.head()
```

```python
def fetch_core(text):
    words = text.split()
    
    if len(words) > 4:
        # Check for 'i3', 'i5', 'i7', 'i9' directly
        if words[4] in ['i3', 'i5', 'i7', 'i9']:
            return words[4]

    if len(words) > 1:
        if words[1] == 'M2':
            return 'i9'
        elif words[1] == 'M1':
            return 'i7'
        elif words[1] == 'Intel':
            return 'i3'
    
    if '5' in text:
        return 'i5'
    elif '7' in text:
        return 'i7'
    elif '9' in text:
        return 'i9'
    
    return 'i3'

df1['processors_core'] = df1['processor'].apply(fetch_core)
df1.head()
```

## LINEAR-REGRESSION

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

step1 = ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(sparse_output=False,drop='first', handle_unknown='ignore'),[0,1,8,11,12])],remainder='passthrough')
step2 = LinearRegression()
pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,Y_train)

Y_pred = pipe.predict(X_test)

print('R2 Score',r2_score(Y_test,Y_pred))
print('Mean Absolute Error',mean_absolute_error(Y_test,Y_pred))
```

## VOTING-REGRESSOR

```python
from sklearn.ensemble import VotingRegressor,StackingRegressor

step1 = ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(sparse_output=False,drop='first', handle_unknown='ignore'),[0,1,8,11,12])],remainder='passthrough')

et= ExtraTreesRegressor(n_estimators=300, max_depth=17, random_state=3,max_samples=.4,max_features=.78, bootstrap=True)
gbdt= GradientBoostingRegressor(n_estimators=800)
rf=RandomForestRegressor(n_estimators=100, max_depth=17, random_state=3,max_samples=.4,max_features=.78)
lasso=Lasso(alpha=0.001)

step2 = VotingRegressor([('et',et),('gbdt',gbdt),('rf',rf),('lasso',lasso)],weights=[1,10,1,1])

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,Y_train)

Y_pred = pipe.predict(X_test)

print('R2 Score',r2_score(Y_test,Y_pred))
print('Mean Absolute Error',mean_absolute_error(Y_test,Y_pred))
```

## STACKING

```python
from sklearn.ensemble import VotingRegressor,StackingRegressor

step1 = ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(sparse_output=False,drop='first', handle_unknown='ignore'),[0,1,8,11,12])],remainder='passthrough')

estimators= [
            ('rf', RandomForestRegressor(n_estimators=100, max_depth=17, random_state=3,max_samples=.4,max_features=.78)),
            ('gbdt', GradientBoostingRegressor(n_estimators=800)),
            ('et', ExtraTreesRegressor(n_estimators=300, max_depth=17, random_state=3,max_samples=.4,max_features=.78, bootstrap=True))
]

step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=50))

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,Y_train)

Y_pred = pipe.predict(X_test)

print('R2 Score',r2_score(Y_test,Y_pred))
print('Mean Absolute Error',mean_absolute_error(Y_test,Y_pred))
```
